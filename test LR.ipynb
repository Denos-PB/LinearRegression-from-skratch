## Linear Regression Model Testing

### Model Implementation
```python
# Linear regression model from scratch with regularization support
# Features: L1/L2 regularization, early stopping, gradient descent optimization

````python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
from LinearRegression import LinearRegression
from sklearn.metrics import r2_score
````

````python
n_samples = 1000
n_features = 10
````

# Generated synthetic regression dataset
 - 1000 samples, 10 features (5 informative, 5 noise)
 - Train/Test split: 70/30 ratio
 - Noise level: 10

````python
X, y = make_regression(n_samples=n_samples, n_features=n_features,
                       n_informative=5, noise=10, random_state=42)
````

````python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
````

````python
fig = plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], y, color="green", marker=".", s=5, alpha=0.6)
plt.xlabel(f'Feature 1 (X[:, 0])')
plt.ylabel('Target (y)')
plt.title('Scatter Plot: Feature 1 vs Target')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

# Training configuration:
 - Learning rate: 0.01
 - Maximum iterations: 1000
 - Convergence tolerance: 1e-5
 - No regularization (penalty=None)

````python
model = LinearRegression(lr=0.01)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
````

````python
r2 = r2_score(y_test,y_pred)
r2
````

````python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.6, color='blue', s=30)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2, label='Perfect prediction')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title(f'Actual vs Predicted Values\nR = {r2_score(y_test, y_pred):.3f}')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````
